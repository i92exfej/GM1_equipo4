Practica 2 de IAA: Regresi√≥n 


```python
import random as rnd


def inicializar():
  w = []
  w.append(rnd.randint(1, 501) * 0.01)
  w.append(rnd.randint(1, 501) * 0.01)
  pendiente = rnd.randint(1, 501) * 0.01
  ap = 0.01
  return w, pendiente, ap 



def estimar(pendiente,w,px):

  v = px[0] * w[0] + px[1] * w[1] + pendiente

  return v


#Metodo 2
def regresionMet2(pendiente,w,px,py,ap):

  estimado = px[0] * w[0] + px[1] * w[1] + pendiente

  pendiente2 = pendiente + ap * (py - estimado)

  for i in range(0,len(w)):
    w[i] = w[i] + ap * px[i] * (py - estimado)



  return w, pendiente2


def MAE(pendiente,w,x,y):
  error = 0
  serror = 0

  for i in range(0,len(y)):
    estimado = x[i][0] * w[0] + x[i][1] * w[1] + pendiente
    error = y[i] - estimado    
    if error < 0:
       error = error * -1
    #print("error en iteracion ",i,": ",error)
    serror = serror + error

  return (1/i) * serror


def MSE(pendiente,w,x,y):
  error = 0
  serror = 0

  for i in range(0,len(y)):
    estimado = x[i][0] * w[0] + x[i][1] * w[1] + pendiente
    error = y[i] - estimado    
    serror = serror + error**2

  return (1/i) * serror









def main():

  w = []
  pendiente = 0
  ap = 0 #aprendizaje 

  w, pendiente, ap = inicializar()
  print("-----------------------------")
  print("Valores aleatorios iniciales: ")
  print("Peso 1: ",w[0], ", Peso 2: ",w[1],", Pendiente: ",pendiente)
  print("-----------------------------")

  valoresx = []
  valoresy = []


  valoresx.append([1,2])
  valoresx.append([1,3])
  valoresx.append([2,3])
  valoresx.append([2,4])
  valoresx.append([3,2])
  valoresx.append([3,5])
  valoresx.append([4,1])

  valoresy.append(1.03)
  valoresy.append(-1.44)
  valoresy.append(4.53)
  valoresy.append(2.24)
  valoresy.append(13.27)
  valoresy.append(5.62)
  valoresy.append(21.53)




  for i in range(0,200):

    j = rnd.randint(0, len(valoresy)-1) 

    w, pendiente = regresionMet2(pendiente,w,valoresx[j],valoresy[j],ap)

  print("---------------------------------")
  print("Los valores finales son tras ",i," iteraciones: ")
  print("Peso 1: ",w[0], ", Peso 2: ",w[1],", Pendiente: ",pendiente)
  print("---------------------------------")

  errorMAE = MAE(pendiente,w,valoresx,valoresy)
  errorMSE = MSE(pendiente,w,valoresx,valoresy)


  print("---------------------------------")
  print("Los errores son: ")
  print("MAE: ",errorMAE, ", MSE: ",errorMSE)
  print("---------------------------------")

 # xi1 = 1
 # xi2 = 2

 # print("---------------------------------")
 # print("Estimando puntos: ")
 # print("x1: ",xi1, ", x2: ",xi2)
 # print("Estimado = ",estimar(pendiente,w,[xi1,xi2]))
 # print("---------------------------------")






if __name__ == "__main__":
    main()


```
